```python
list_of_list=[]

import numpy as np
for feature_type in df_final.feature_type.unique():
    k1= df_final[df_final["feature_type"]==feature_type].reset_index(drop=True)
    if(len(k1) !=0):
        print(feature_type)
    
    for measurement in k1.measurement.unique():
        k2= k1[k1["measurement"]==measurement].reset_index(drop=True)
        if(len(k2) !=0):
            print("  ", measurement)     
            
        endpoint_counter = 0 
        for endpoint in k2.endpoint.unique():
            k3= k2[k2["endpoint"]==endpoint].reset_index(drop=True)
            if(len(k3) !=0):
                print("    ",endpoint)
                
            cell_process_counter = 0        
            
            for cell_process in k3.cell_process.unique():
                k4= k3[k3["cell_process"]==cell_process].reset_index(drop=True)          
                if(len(k4) !=0):
                    features_from_ontology = set(k4[k4["cell_process"]==cell_process].reset_index()["selected_features"][0])

                    overlap = features_from_task & features_from_ontology
                    cell_process_perc_overlap = float(len(overlap)) / len(features_from_ontology) * 100
                    print("      ",cell_process," ",np.round(cell_process_perc_overlap, 2),"%")
                    
                    if(cell_process_perc_overlap>10):
                        cell_process_counter=cell_process_counter+1
                    row=[feature_type, measurement, endpoint, cell_process, cell_process_perc_overlap]   
                    list_of_list.append(row)
                    
```

```python
cp_ontology_enrichment= pd.DataFrame(list_of_list, columns=[["feature_type", "measurement", "endpoint", "cell_process", "cell_process_perc_overlap"]])
cp_ontology_enrichment.to_csv("Trial_explanations.csv", index=False)
cp_ontology_enrichment = pd.read_csv("Trial_explanations.csv")
cp_ontology_enrichment
```
