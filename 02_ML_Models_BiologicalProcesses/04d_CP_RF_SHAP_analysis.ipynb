```python
import pandas as pd
```

```python
df = pd.read_csv("metrics_df_CP_RF_preds_SHAP_allendpoints_singlerun_withnames.csv")
df
```

```python
from ast import literal_eval
import numpy as np

df.Filtered_CP_SHAP_features = df.Filtered_CP_SHAP_features.apply(literal_eval)
df
```

```python
import Check_Ontology
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
list_of_lists_a=[]

for labels_name in df.labels_name.unique():
    
    test_compounds = df[df["labels_name"]==labels_name].reset_index(drop=True)  
    print(labels_name)
    
    for InChICode_standardised in test_compounds.InChICode_standardised.to_list():
    
        common_name = test_compounds[test_compounds["InChICode_standardised"]==InChICode_standardised]["Common_name"].values[0]
        print(common_name)
        
        shap_imp_features = set(test_compounds[test_compounds["InChICode_standardised"]==InChICode_standardised].reset_index()["Filtered_CP_SHAP_features"][0])

        ontology_enrichment = Check_Ontology.return_ontology(shap_imp_features)

        endpoint_wise = ontology_enrichment.groupby(["endpoint"]).mean()[ontology_enrichment.groupby(["endpoint"]).mean()["cell_process_perc_overlap"]> 5]
        endpoint_wise = endpoint_wise.sort_values(by =["cell_process_perc_overlap"], ascending = False).reset_index()

        if(len(endpoint_wise)>0):  
            sns.barplot(data=endpoint_wise, x="cell_process_perc_overlap", y="endpoint",color="grey")
            #plt.gcf().set_dpi(100)
            plt.show()
            
        try:
            first = endpoint_wise.iloc[0].endpoint
            first_value = endpoint_wise.iloc[0].cell_process_perc_overlap

        except:
            print("")

        try:
            second = endpoint_wise.iloc[1].endpoint
            second_value = endpoint_wise.iloc[1].cell_process_perc_overlap            
        
        except:
            print("")
            
        row=[common_name, labels_name, first, first_value, second, second_value]
        list_of_lists_a.append(row)
            
```

```python
phenotype_per_compound= pd.DataFrame(list_of_lists_a, columns=["common_name", "labels_name", "firsthealth_phenotype", "firsthealth_phenotype_value", "secondhealth_phenotype", "secondhealth_phenotype_value"])
phenotype_per_compound
```

```python
phenotype_per_compound.to_csv("cell_health_phenotype_wise_percompound.csv", index=False)
```

```python
import numpy as np

for label in df.labels_name.to_list():
    
    if(df[df["labels_name"]==label]["AUC"].values <=0.50):
        continue
    
    print(label, "*****************************")
    
    ontology_enrichment = Check_Ontology.return_ontology(set(df[df["labels_name"]==label].reset_index()["Filtered_CP_imp_features"][0]))
    
    for feature_type in ontology_enrichment.feature_type.unique():
        k1= ontology_enrichment[ontology_enrichment["feature_type"]==feature_type].reset_index(drop=True)
        print(feature_type,"* ",np.round(len(k1[k1["cell_process_perc_overlap"]>=10])/len(k1) *100, 2), "%")

        for measurement in k1.measurement.unique():
            k2= k1[k1["measurement"]==measurement].reset_index(drop=True)
            print("  ", measurement," ", np.round(len(k2[k2["cell_process_perc_overlap"]>=10])/len(k2) *100, 2), "%")

            for end in k2.endpoint.unique():
                k3= k2[k2["endpoint"]==end].reset_index(drop=True)
                print("$$$    ",end," ", np.round(len(k3[k3["cell_process_perc_overlap"]>=10])/len(k3) *100, 2), "%")

                for cell_process in k3.cell_process.unique():
                    k4= k3[k3["cell_process"]==cell_process].reset_index(drop=True)          
                    print("%%%%      ",cell_process," ",np.round(float(k4["cell_process_perc_overlap"]), 2),"%")
```
